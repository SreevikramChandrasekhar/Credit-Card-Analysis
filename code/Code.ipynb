```python
rep_metrics_df = rep_metrics_df[['Customer','Product','City','Year','Amount']]
rep_metrics_df
```

```python
#Function that takes product and year as input
#Output is top 5 customers in terms of repayment in every city based on input parameters
def rep_metrics(df, prod, y):
    
    filtered_df = df[rep_metrics_df['Product'] == prod]
    filtered_df = filtered_df[filtered_df['Year'] == y]

    filtered_df = filtered_df.groupby(['Product','Year','City','Customer'])['Amount'].sum().reset_index()
    filtered_df.sort_values(by = ['City','Amount'], ascending = [True,False], inplace = True)
    # print(filtered_df)
    top10_cust = filtered_df.groupby('City').head(5)
    print(top10_cust)
    
```
